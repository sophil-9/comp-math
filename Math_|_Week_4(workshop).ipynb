```python
#Task 1
import matplotlib.pyplot as plt
import random
x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]
colors = ["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]
plt.scatter(x, y, c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

```python
#Task 2-b
import matplotlib.pyplot as plt

y1=[8,150,80]
y2=[54,77,54]
y3=[93,32,100]
y4=[116,11,76]
y5=[137,6,93]
y6=[184,1,72]

g=['bear','dolphins','whales']

data=[y1,y2,y3,y4,y5,y6]
years=[2017,2018,2019,2020,2021,2022]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.pie(data[i],labels=g)
    plt.title(years[i])
plt.tight_layout()
plt.legend()
plt.show()
```
